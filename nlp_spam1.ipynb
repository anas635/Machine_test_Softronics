```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('/content/drive/MyDrive/ML datasets/Unsupervised and NLP/spam_ham_dataset.csv')
df
```

```python
df.isna().sum(),df.duplicated().sum()
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from string import punctuation
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')

stop=stopwords.words('english')
lemma=WordNetLemmatizer()
def preprocess(sent):
  a=word_tokenize(''.join([i for i in sent.lower() if i not in punctuation]))
  return ' '.join(lemma.lemmatize(i,'v') for i in a if i not in stop)
```

```python
tekst="I love programming in Python! It's amazing, and I learn new things every day."
preprocess(tekst)
```

```python
df['text']=df['text'].apply(func=preprocess)
```

```python
df
```

```python
x=df['text']
y=df['label_num']
```

```python
df['label_num'].value_counts()
```

count vectoriser

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
vectoriser=CountVectorizer()
vectoriser1=TfidfVectorizer()
x_vector=vectoriser.fit_transform(x)
x_vector1=vectoriser1.fit_transform(x)
```

```python
x_vector.shape,x_vector1.shape
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x_vector1,y,test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from xgboost import XGBClassifier
models=classification_algorithms = [
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    GradientBoostingClassifier(),
    XGBClassifier(),
    SVC(),
    MultinomialNB(),
    KNeighborsClassifier(),
]
```

```python
from sklearn.metrics import accuracy_score,classification_report
for model in models:
  model.fit(xtrain,ytrain)
  ypred=model.predict(xtest)
  print(f'{model}:= \n {classification_report(ytest,ypred)}')
  print('==================================================')
```

Sampling

```python
from imblearn.over_sampling import RandomOverSampler
sampler=RandomOverSampler()
x_res,y_res=sampler.fit_resample(x_vector1,y)
```

```python
y_res.value_counts()
```

```python
xrtrain,xrtest,yrtrain,yrtest=train_test_split(x_res,y_res,test_size=0.3)
```

```python
for model in models:
  model.fit(xrtrain,yrtrain)
  yrpred=model.predict(xrtest)
  print(f'{model}:= \n {classification_report(yrtest,yrpred)}')
  print('==================================================')
```
